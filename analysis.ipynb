```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('Amazon Sales data.csv')

# Display the DataFrame
print(df)
```

```python
# Step 3: Data Cleaning
# Example: Convert dates to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Example: Handle missing values (if any)
df.dropna(inplace=True)  # Drop rows with missing values
```

```python
# Step 4: Data Transformation and Feature Engineering
# Example: Calculate Total Profit
df['Total Profit'] = df['Total Revenue'] - df['Total Cost']
```

```python
# Step 5: Data Analysis
# Example: Month-wise Sales Trend
monthly_sales = df.groupby(df['Order Date'].dt.month)['Units Sold'].sum()

# Example: Year-wise Sales Trend
yearly_sales = df.groupby(df['Order Date'].dt.year)['Units Sold'].sum()

# Example: Profitability Analysis
profitable_items = df[df['Total Profit'] > 0]
```

```python
# Step 6: Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Example: Visualize Month-wise Sales Trend
plt.figure(figsize=(10, 6))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker='o')
plt.title('Month-wise Sales Trend')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.grid(True)
plt.show()
```

```python
# Calculate monthly sales
monthly_sales = df.groupby(df['Order Date'].dt.month)['Units Sold'].sum()

# Plotting the month-wise sales trend
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-')
plt.title('Month-wise Sales Trend')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

```python
# Calculate yearly sales
yearly_sales = df.groupby(df['Order Date'].dt.year)['Units Sold'].sum()

# Plotting the year-wise sales trend
plt.figure(figsize=(10, 6))
plt.plot(yearly_sales.index, yearly_sales.values, marker='o', linestyle='-')
plt.title('Year-wise Sales Trend')
plt.xlabel('Year')
plt.ylabel('Units Sold')
plt.grid(True)
plt.show()
```

```python
# Calculate yearly month-wise sales
yearly_monthly_sales = df.groupby([df['Order Date'].dt.year, df['Order Date'].dt.month])['Units Sold'].sum()

# Plotting the yearly month-wise sales trend
plt.figure(figsize=(12, 8))
for year in yearly_monthly_sales.index.levels[0]:
    sales_data = yearly_monthly_sales.loc[year]
    plt.plot(sales_data.index, sales_data.values, marker='o', label=f'Year {year}')

plt.title('Yearly Month-wise Sales Trend')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend(title='Year')
plt.grid(True)
plt.show()
```

```python
# Calculate yearly month-wise sales
yearly_monthly_sales = df.groupby([df['Order Date'].dt.year, df['Order Date'].dt.month])['Units Sold'].sum()

# Iterate over each year and plot
for year in yearly_monthly_sales.index.levels[0]:
    # Filter data for the current year
    sales_data = yearly_monthly_sales.loc[year]
    
    # Plotting the yearly month-wise sales trend for the current year
    plt.figure(figsize=(10, 6))
    plt.plot(sales_data.index, sales_data.values, marker='o', linestyle='-', label=f'Year {year}')
    plt.title(f'Year {year} Month-wise Sales Trend')
    plt.xlabel('Month')
    plt.ylabel('Units Sold')
    plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Calculate total units sold per month across all years
monthly_total = df.groupby(df['Order Date'].dt.month)['Units Sold'].sum()

# Plotting the bar chart for monthly sales
plt.figure(figsize=(10, 6))
plt.bar(monthly_total.index, monthly_total.values, tick_label=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.grid(True)
plt.show()
```

```python
# Example analysis to calculate and visualize some key metrics
import matplotlib.pyplot as plt
import seaborn as sns

# Example: Calculate total revenue, cost, and profit
df['Total Revenue'] = df['Units Sold'] * df['Unit Price']
df['Total Cost'] = df['Units Sold'] * df['Unit Cost']
df['Total Profit'] = df['Total Revenue'] - df['Total Cost']

# Example: Visualize total revenue and profit over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Order Date', y='Total Revenue', data=df, label='Total Revenue')
sns.lineplot(x='Order Date', y='Total Profit', data=df, label='Total Profit')
plt.title('Total Revenue and Profit Over Time')
plt.xlabel('Order Date')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)
plt.show()

# Example: Scatter plot for units sold vs. unit price
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Units Sold', y='Unit Price', data=df, hue='Item Type', palette='Set2', alpha=0.7)
plt.title('Units Sold vs. Unit Price')
plt.xlabel('Units Sold')
plt.ylabel('Unit Price')
plt.legend()
plt.grid(True)
plt.show()

# Example: Correlation heatmap
corr_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is your DataFrame with Amazon sales data loaded
# Example: df = pd.read_csv('amazon_sales_data.csv')

# Convert 'Order Date' to datetime format if it's not already
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Extract month and year from 'Order Date'
df['Order Month'] = df['Order Date'].dt.month
df['Order Year'] = df['Order Date'].dt.year

# Calculate total revenue per item type per month
df['Total Revenue'] = df['Units Sold'] * df['Unit Price']
monthly_revenue = df.groupby(['Item Type', 'Order Year', 'Order Month'])['Total Revenue'].sum().reset_index()

# Display the first few rows of the dataframe
monthly_revenue.head()
```

```python
# Calculate KPIs
total_sales = df['Total Revenue'].sum()
average_order_value = df['Total Revenue'].mean()
total_profit = df['Total Profit'].sum()
profit_margin = (total_profit / total_sales) * 100 if total_sales != 0 else 0

# Display KPIs
print(f"Total Sales: ${total_sales:.2f}")
print(f"Average Order Value: ${average_order_value:.2f}")
print(f"Total Profit: ${total_profit:.2f}")
print(f"Profit Margin: {profit_margin:.2f}%")
```

```python
# Calculate KPIs
total_sales = df['Total Revenue'].sum()
average_order_value = df['Total Revenue'].mean()
total_profit = df['Total Profit'].sum()
profit_margin = (total_profit / total_sales) * 100 if total_sales != 0 else 0

# Round values for better readability
total_sales_rounded = round(total_sales, 2)
average_order_value_rounded = round(average_order_value, 2)
total_profit_rounded = round(total_profit, 2)
profit_margin_rounded = round(profit_margin, 2)

# Display KPIs
print(f"Total Sales: ${total_sales_rounded:.2f}")
print(f"Average Order Value: ${average_order_value_rounded:.2f}")
print(f"Total Profit: ${total_profit_rounded:.2f}")
print(f"Profit Margin: {profit_margin_rounded:.2f}%")
```

```python
import matplotlib.pyplot as plt

# Plotting KPIs
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Function to add border around text
def add_border_around_text(ax, x, y, text, fontsize=16, color='white'):
    bbox_props = dict(boxstyle='round,pad=0.3', fc='black', ec='black', lw=2)
    ax.text(x, y, text, ha='center', va='center', fontsize=fontsize, color=color, bbox=bbox_props)

# Total Sales
axs[0, 0].bar("Total Sales", total_sales, color='blue')
add_border_around_text(axs[0, 0], 0.5, total_sales * 0.5, f"${total_sales_rounded:.2f}")

axs[0, 0].set_title('Total Sales')

# Average Order Value
axs[0, 1].bar("Average Order Value", average_order_value, color='green')
add_border_around_text(axs[0, 1], 0.5, average_order_value * 0.5, f"${average_order_value_rounded:.2f}")

axs[0, 1].set_title('Average Order Value')

# Total Profit
axs[1, 0].bar("Total Profit", total_profit, color='orange')
add_border_around_text(axs[1, 0], 0.5, total_profit * 0.5, f"${total_profit_rounded:.2f}")

axs[1, 0].set_title('Total Profit')

# Profit Margin
axs[1, 1].bar("Profit Margin", profit_margin, color='red')
add_border_around_text(axs[1, 1], 0.5, profit_margin * 0.5, f"{profit_margin_rounded:.2f}%")

axs[1, 1].set_title('Profit Margin')

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```
